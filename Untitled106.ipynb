```python
# =========================================
# GWR COVID-19 (COLAB VERSION)
# =========================================

import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from mgwr.gwr import GWR
from mgwr.sel_bw import Sel_BW
import numpy as np
import os

try:
    print("Loading data...")

    # ===============================
    # 0. PATH (COLAB)
    # ===============================
    data_path = "/content/covid19_eng.gpkg"
    output_dir = "/content/figs"
    os.makedirs(output_dir, exist_ok=True)

    gdf = gpd.read_file(data_path)

    # ===============================
    # 1. DEFINE VARIABLES
    # ===============================

    # Dependent variable
    gdf['covid19_r'] = (gdf['X2020.04.14'] / gdf['Residents']) * 100000

    # Independent variables
    gdf['imd'] = gdf['IMD...Average.score']
    gdf['lt_illness'] = gdf['Long_term_ill'] / gdf['Residents']
    gdf['crowded'] = gdf['Crowded_housing'] / gdf['Households']

    # Clean data
    gdf = gdf.replace([np.inf, -np.inf], np.nan)
    gdf = gdf.fillna(0)

    gdf_clean = gdf[
        ['covid19_r', 'imd', 'lt_illness', 'crowded', 'geometry']
    ].copy()

    # ===============================
    # 2. GWR INPUT
    # ===============================

    X = gdf_clean[['imd', 'lt_illness', 'crowded']].values
    y = gdf_clean['covid19_r'].values.reshape(-1, 1)

    coords = list(zip(
        gdf_clean.geometry.centroid.x,
        gdf_clean.geometry.centroid.y
    ))

    # ===============================
    # 3. BANDWIDTH SELECTION
    # ===============================

    print("Selecting bandwidth...")
    bw_selector = Sel_BW(
        coords, y, X,
        kernel='bisquare',
        fixed=False
    )
    bw = bw_selector.search(criterion='AICc')
    print(f"Optimal bandwidth: {bw}")

    # ===============================
    # 4. FIT GWR
    # ===============================

    print("Fitting GWR...")
    gwr_model = GWR(
        coords, y, X, bw,
        kernel='bisquare',
        fixed=False
    )
    gwr_results = gwr_model.fit()

    # ===============================
    # 5. EXTRACT RESULTS
    # ===============================

    gdf_clean['gwr_imd'] = gwr_results.params[:, 1]
    gdf_clean['gwr_lt_illness'] = gwr_results.params[:, 2]
    gdf_clean['gwr_crowded'] = gwr_results.params[:, 3]

    gdf_clean['gwr_t_imd'] = gwr_results.tvalues[:, 1]
    gdf_clean['gwr_t_lt_illness'] = gwr_results.tvalues[:, 2]
    gdf_clean['gwr_t_crowded'] = gwr_results.tvalues[:, 3]

    # ===============================
    # 6. PLOT FUNCTION
    # ===============================

    def plot_significant_map(col, t_col, title, fname):
        fig, ax = plt.subplots(1, 1, figsize=(10, 10))

        gdf_clean.plot(
            ax=ax,
            color='lightgrey',
            edgecolor='white'
        )

        sig = abs(gdf_clean[t_col]) > 1.96

        if sig.sum() > 0:
            gdf_clean[sig].plot(
                column=col,
                ax=ax,
                cmap='viridis',
                scheme='quantiles',
                legend=True,
                legend_kwds={'title': title, 'loc': 'lower right'}
            )

        ax.set_title(f"{title} (|t| > 1.96)")
        ax.axis('off')

        plt.savefig(f"{output_dir}/{fname}", dpi=300, bbox_inches='tight')
        plt.close()
        print(f"Saved: {fname}")

    # ===============================
    # 7. GENERATE MAPS
    # ===============================

    print("Generating maps...")
    plot_significant_map(
        'gwr_imd',
        'gwr_t_imd',
        'Koefisien IMD (Deprivation)',
        'covid_gwr_coef_imd_sig.png'
    )

    plot_significant_map(
        'gwr_lt_illness',
        'gwr_t_lt_illness',
        'Koefisien Penyakit Jangka Panjang',
        'covid_gwr_coef_lt_illness_sig.png'
    )

    plot_significant_map(
        'gwr_crowded',
        'gwr_t_crowded',
        'Koefisien Kepadatan Hunian',
        'covid_gwr_coef_crowded_sig.png'
    )

    print("DONE ✅")

except Exception as e:
    print("ERROR ❌")
    import traceback
    traceback.print_exc()
```

```python
def plot_variable_map(var, title):
    fig, ax = plt.subplots(1,1, figsize=(8,8))
    gdf_clean.plot(
        column=var,
        ax=ax,
        legend=True,
        cmap='viridis',
        scheme='quantiles'
    )
    ax.set_title(title)
    ax.axis('off')
    plt.show()

plot_variable_map('covid19_r', 'Sebaran COVID-19 Rate')
plot_variable_map('imd', 'Sebaran IMD')
plot_variable_map('lt_illness', 'Sebaran Penyakit Jangka Panjang')
plot_variable_map('crowded', 'Sebaran Kepadatan Hunian')
```

```python
import statsmodels.api as sm

X_ols = gdf_clean[['imd', 'lt_illness', 'crowded']]
X_ols = sm.add_constant(X_ols)
y_ols = gdf_clean['covid19_r']

ols_model = sm.OLS(y_ols, X_ols).fit()
print(ols_model.summary())
```

```python
from libpysal.weights import Queen
from esda.moran import Moran

# Spatial weights
w = Queen.from_dataframe(gdf_clean)
w.transform = 'r'

# Residual OLS
residuals = ols_model.resid

# Moran's I
moran = Moran(residuals, w)

print("Moran's I:", moran.I)
print("p-value:", moran.p_sim)
```
